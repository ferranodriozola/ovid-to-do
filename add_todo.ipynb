```python
import json

text = """
Transcriure Puteolanus
(X) *Dedicatòria
*Vida
*Índex

Omplir excel Puteolanus
Corregir xml Puteolanus
Enviar a Pere Comitibus i Puteolanus + qüestions
*Comitibus: què ha canviat
*mirar correu i whatsapp que li havia enviat

Creador xml personatges
*csv persones drive
*csv llocs drive
*db drive (occupation)
*db drive (certainty)
*gestionar db en línea amb el generador xml normal en local:
*db drive (languages) (!!)
*db drive (city) (!!)
*db drive (country) (!!)
*fer python persones
*fer python llocs
"""


def transforma_linies_a_todo(text: str) -> str:
    # Omet línies buides, detecta '(X) ' i serialitza el text per generar codi Python vàlid.
    linies = [linia.strip() for linia in text.splitlines() if linia.strip()]

    def converteix_linia(linia: str) -> str:
        is_done = linia.startswith("(X) ")
        linia_neta = linia[4:].lstrip() if is_done else linia
        done_part = ", is_done=True" if is_done else ""
        return f"        Todo(text={json.dumps(linia_neta, ensure_ascii=False)}{done_part}),"

    return "\n".join(converteix_linia(linia) for linia in linies)


print(transforma_linies_a_todo(text))
```
